*Below here are the codes of how I perform "Exploratory Data Analysis" using `pandas`, `matplotlib`, and `seaborn` libraries*

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
pd.set_option('display.max_columns', None)
plt.rcParams['figure.figsize'] = (4, 2.5)
plt.rcParams.update({'font.size': 7})
```

```python
# read 'Merged Data'
df = pd.read_excel('dataset/Merged Data.xlsx').drop(['Unnamed: 0'], axis = 1)
display(df.head(), df.shape)
```

```python
# analyze the content categories
categories = df.groupby('Category')\
                .agg(total_score = ('Score', 'sum'),
                     total_reaction = ('Score', 'count'), 
                     total_post = ('Content ID', 'nunique'))\
                .rename(columns = {'total_score':'Total Score', 'total_reaction':'Total Reaction', 'total_post':'Total Post'})\
                .sort_values('Total Score', ascending = False)                   
categories.index = [' '.join([word.capitalize() for word in ind.split(' ')]) for ind in categories.index]
categories.index.name = 'Category'
categories.reset_index(inplace = True)
categories
```

```python
# find the correlation between 'Total Score', 'Total Reaction', 'Total Post'
correlation = categories.select_dtypes(exclude = 'object').corr()

fig, ax = plt.subplots()
mask = np.triu(np.ones_like(correlation))
sns.heatmap(correlation, cmap = 'Purples', vmin = 0, vmax = 1, annot = True, ax = ax, mask = mask)
ax.set_yticklabels(correlation.index, rotation = 0)
ax.set(title = 'Pearson Correlation Matrix')
plt.show()
fig.savefig('image/Correlation Matrix.png', bbox_inches = 'tight', dpi = 300)
```

```python
fig, ax = plt.subplots(figsize = (9, 5), ncols = 3, sharey = True)
for i, col in enumerate(['Total Score', 'Total Reaction', 'Total Post']):
    sns.barplot(data = categories, 
                x = col, y = 'Category', hue = 'Category', palette = 'tab20', ax = ax[i], alpha = 0.25)
    sns.barplot(data = categories.head(5), 
                x = col, y = 'Category', hue = 'Category', palette = 'tab20', ax = ax[i])
ax[0].set(title = 'by Total Popularity Scores', xlabel = 'Popularity Scores', ylabel = None)
ax[1].set(title = 'by Total Reactions', xlabel = 'Reactions', ylabel = None)
ax[2].set(title = 'by Total Posts', xlabel = 'Posts', ylabel = None)
ax[0].set_xlim([categories['Total Score'].min() - 1000, categories['Total Score'].max() + 1000])
ax[1].set_xlim([categories['Total Reaction'].min() - 100, categories['Total Reaction'].max() + 100])
ax[2].set_xlim([categories['Total Post'].min() - 2, categories['Total Post'].max() + 2])
plt.suptitle('Content Categories in Social Buzz', ha = 'center', size = 11)
plt.show()

fig.savefig('image/Content by Category.png', bbox_inches = 'tight', dpi = 300)
```

```python
top5 = categories.head(5)

fig, ax = plt.subplots(figsize = (7, 3))
top5.plot(explode = [0.1, 0, 0, 0, 0], kind = 'pie', y = 'Total Score', autopct = '%.2f%%', colors = sns.color_palette('tab20'), labels = top5.Category, legend = None, ax = ax)
ax.set(title = 'Top 5 Most Popular Categories', xlabel = None, ylabel = None)
plt.show()

fig.savefig('image/Top 5 Contents.png', bbox_inches = 'tight', dpi = 300)
```
